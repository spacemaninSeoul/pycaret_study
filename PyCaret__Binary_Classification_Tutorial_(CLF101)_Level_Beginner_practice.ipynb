# 2. 튜토리얼 데이터셋

튜토리얼에서는 UCI에서 얻을 수 있는 데이터셋인 'Default of Credit Card Clients Dataset'을 이용한다. 이 데이터셋은 2005년 4월부터 9월까지 타이완의 신용 카드 고객에 대한 default payments, demographic factors, credit data, payment history, and billing statements가 담겨있다.

 - ID: ID of each client
 - LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit)
 - SEX: 성별 (1=male, 2=female)
 - EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
 - MARRIAGE: Marital status (1=married, 2=single, 3=others)
 - AGE: Age in years
 - PAY_0 to PAY_6: n달 간의 상환 상태 (PAY_0 = last month ... PAY_6 = 6 months ago) (Labels: -1= 기한내 지불, 1=payment delay for one month, 2=payment delay for two months, ... 8=payment delay for eight months, 9=payment delay for nine months and above)
 - BILL_AMT1 to BILL_AMT6: Amount of bill statement by n months ago ( BILL_AMT1 = last_month .. BILL_AMT6 = 6 months ago)
 - PAY_AMT1 to PAY_AMT6: Amount of payment by n months ago ( BILL_AMT1 = last_month .. BILL_AMT6 = 6 months ago)

 - default: Default payment (1=yes, 0=no)Target Column


Dataset Acknowledgement:
Lichman, M. (2013). UCI Machine Learning Repository. Irvine, CA: University of California, School of Information and Computer Science.



# 3. 데이터 가져오기

```python
from pycaret.datasets import get_data
```

```python
dataset = get_data('credit')
```

```python
# 데이터의 크기를 확인한다
dataset.shape
```

In order to demonstrate the **predict_model()** function on unseen data, a sample of 1200 records has been withheld from the original dataset to be used for predictions. This should not be confused with a train/test split as this particular split is performed to simulate a real life scenario. Another way to think about this is that these 1200 records are not available at the time when the machine learning experiment was performed.

predict_model() 기능을 unseen data에 시현하기 위해, 샘플에서 1200개의 데이터를 제거한다.



```python
data = dataset.sample(frac=0.95, random_state=786)
data_unseen = dataset.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)

print('Data for Modeling: ', str(data.shape))
print('Unseen Data For Predictions: ', str(data_unseen.shape))
```

# 4. PyCaret 환경 설정하기

**setup()** 함수를 통해 pycaret에서 모델링을 할 수 있다. setup()은 다른 기능을 사용하기 전에 반드시 먼저 수행되어야 한다. 의무적으로 적시해야할 두 개의 파라미터가 있는데, data와 target 파라미터이다. 그외 파라미터들은 선택적이다.

 - data : 모델에 사용할 데이터를 지정한다.
 - target : 타겟값으로 사용할 피처를 지정한다.
 - train_size : 데이터에서 학습에 사용할 데이터의 비율을 지정한다.
 - session_id : 사이킷런의 random_state와 같은 역할을 한다.



```python
data.head()
```

```python
from pycaret.classification import *
exp_clf101 = setup(data = data, target ='default', session_id =123)
```

 - session_id : 나중에 재현할 수 있도록 하는 pseudo-random number seed이다.
